### 2. Create a function that accepts two arguments, an integer, and a vector of integers. It returns the count of the number of the occurrences of the integer in the input vector.

```python
def countOccurrences(arr, n, x):
    res = 0
    for i in range(n):
        if x == arr[i]:
            res += 1
    return res
```

```python
arr = [1, 2, 2, 2, 2, 3, 4, 7 ,8 ,8]
n = len(arr)
x = 4
print (countOccurrences(arr, n, x))
```

### 4. Create a function that will return TRUE if an input integer is prime number. otherwise retuen false

```python
n=int(input("Enter number: ")) 
k=0 
for i in range(2,n//2+1): 
    if(n%i==0): 
        k=k+1 
if(k<=0): 
    print("True") 
else: 
    print("False") 
```
